Data PreProcessing

```python
df=pd.read_csv('C:\main\Desktop\ML\Main\data\health_dataset.csv')
```

```python
df
```

```python
# remove unnnamed column
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

```python
df = df.fillna('')
df
```

```python

all_symptoms = set()
for row in df.iloc[:, 1:].values:
    all_symptoms.update(row)


```

```python
# Remove empty strings from the symptom list
all_symptoms.discard('')

```

```python
# Initialize a new DataFrame for one-hot encoding
one_hot_df = pd.DataFrame(0, index=df.index, columns=list(all_symptoms))
one_hot_df
```

```python
# Fill the DataFrame with 1s where symptoms exist
for index, row in df.iterrows():
    for symptom in row[1:]:  # Symptoms start from column index 1
        if symptom:
            one_hot_df.loc[index, symptom] = 1


```

```python
# Add the 'Disease' column back
one_hot_df.insert(0, 'Disease', df['Disease'])
```

```python
one_hot_df
```

```python
one_hot_df.rename(columns={'Disease': 'diseases'}, inplace=True)
```

```python
one_hot_df.to_csv("C:\main\Desktop\ML\Main\data\dataset_cleaned.csv", index=False)
```

```python
import pandas as pd



# Separate features and target
X = one_hot_df.drop(columns=["diseases"])
y = one_hot_df["diseases"]

# Combine symptoms with the label
df_combined = X.copy()
df_combined["diseases"] = y

# Group by disease and sum symptom columns
symptom_counts = df_combined.groupby("diseases").sum()

# Count non-zero symptom columns (i.e., symptoms present in at least 1 case of the disease)
symptoms_per_disease = (symptom_counts > 0).sum(axis=1)

print("Symptoms per disease:")
print(symptoms_per_disease.sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt

symptoms_per_disease.sort_values(ascending=True).plot(kind='barh', figsize=(10, 8), color='skyblue')
plt.title('Number of Unique Symptoms per Disease')
plt.xlabel('Number of Symptoms')
plt.ylabel('Disease')
plt.tight_layout()
plt.show()
```
